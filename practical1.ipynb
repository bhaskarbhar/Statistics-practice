```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statistics

%matplotlib inline
```

```python
#mean, median, mode
df = sns.load_dataset("tips")
df.head()
```

```python
print(np.mean(df["total_bill"]))
print(np.median(df["total_bill"]))
print(statistics.mode(df["total_bill"]))
```

```python
#boxplot
sns.boxplot(df["total_bill"]) #I can see the outliers here
```

```python
#histogram
sns.histplot(df["total_bill"],kde=True)
#As we can see below the distribution is not normally disrtibuted
```
